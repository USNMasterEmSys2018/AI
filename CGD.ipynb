```python

from mpl_toolkits.mplot3d import Axes3D
import ploydataset
import numpy
#initialization
data = ploydataset.ploydataset(2)
data.generatedata()
  
X, J = ploydataset.fetchdesentdataset(data.var,data.y)
#plot3d(data.var,fe, data.y)

print("z= a0 + a1*x")
```

```python
step=10
rate = 0.00005
max_iters = 100000
iters = 0
precision = 0.000001
step_size =2
if(data.feature[0]>0):
    fe = [-3]*2
else:
    fe = [3]*2
feture = [fe]*1
j, dj, h = ploydataset.f(fe,data.var,data.y)

while(step_size > precision and iters < max_iters):
    prev_fe = fe
    j, dj, h = ploydataset.f(fe, data.var, data.y)
    fe = fe - rate * (dj) #Grad descent
    step_size = 0
    
    for i in range(len(fe)):
        step_size = step_size + abs(fe[i]-prev_fe[i])  
        
    iters = iters+1 #iteration count
    if iters%step==0:
        ploydataset.plotfigure(feture, data.var, data.y)
        feture.append(fe)
        step = int(input("please input training cycle you want to skip: "))
        if step==0:
            step = 10
print("Training finished")
```
