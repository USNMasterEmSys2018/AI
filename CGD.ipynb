```python

from mpl_toolkits.mplot3d import Axes3D
import ploydataset
#initialization
data = ploydataset.ploydataset(2)
data.generatedata()
  
X, J = ploydataset.fetchdesentdataset(data.var,data.y)
fe = [0]*2
j, dj, h = ploydataset.f(fe,data.var,data.y)
#plot3d(data.var,fe, data.y)
step=10
rate = 0.00005
max_iters = 100000
iters = 0
precision = 0.000001
step_size =2
feture = [[0]*2]
print("z= a0 + a1*x")
```

```python

while(step_size > precision and iters < max_iters):
    prev_fe = fe
    j, dj, h = ploydataset.f(fe, data.var, data.y)
    fe = fe - rate * (dj) #Grad descent
    step_size = 0
    
    for i in range(len(fe)):
        step_size = step_size + abs(fe[i]-prev_fe[i])  
        
    iters = iters+1 #iteration count
    if iters%step==0:
        ploydataset.plotfigure(fe, data.var, data.y)
        feture.append(fe)
```

```python
ploydataset.plot3d(data.var,feture, data.y, X, J) 
```
